# Imports

```python
import bs4
import requests
import re
import matplotlib.pyplot as plt
```

## Solution for ex 1

```python
url = 'https://www.elgiganten.dk/catalog/gaming/dk-gaming-laptop/gaming-laptop'
r = requests.get(url)
r.raise_for_status()
soup = bs4.BeautifulSoup(r.text, 'html.parser')
price_list_div = soup.select('.product-list-container .product-price')
name_list_div = soup.select('.product-list-container .product-name')

my_dict = {}
for idx in range(0, len(name_list_div)):
    my_dict[name_list_div[idx].getText()] = price_list_div[idx].getText()

print(my_dict)
```

## Solution for ex 2

```python
gp = re.compile(r'(Asus).([\w ]+)*')
gp.findall(str(my_dict.keys()))
```

## Solution for ex 3

```python
sorted_dict = {k: v for k, v in sorted(my_dict.items(), key=lambda item: item[1])}
plt.bar(sorted_dict.keys(), sorted_dict.values())
plt.xticks(rotation=90)
plt.figure(figsize=(20,10))
```
